```python
#crop image
height_offset = 120
width_offset = 370

image1pos = (70, 180)
image2pos = (295, 180)
image3pos = (525, 180)
image4pos = (760, 180)
image5pos = (980, 180)

image1pos_downsampled = (image1pos[0]//2, image1pos[1]//2) 
image2pos_downsampled = (image2pos[0]//2, image2pos[1]//2)
image3pos_downsampled = (image3pos[0]//2, image3pos[1]//2)
image4pos_downsampled = (image4pos[0]//2, image4pos[1]//2)
image5pos_downsampled = (image5pos[0]//2, image5pos[1]//2)


image = cv2.imread("../../bottomFrame.png")



image1 = image[image1pos[0]:image1pos[0] + height_offset, image1pos[1]:image1pos[1] + width_offset]
image2 = image[image2pos[0]:image2pos[0] + height_offset, image2pos[1]:image2pos[1] + width_offset]
image3 = image[image3pos[0]:image3pos[0] + height_offset, image3pos[1]:image3pos[1] + width_offset]
image4 = image[image4pos[0]:image4pos[0] + height_offset, image4pos[1]:image4pos[1] + width_offset]
image5 = image[image5pos[0]:image5pos[0] + height_offset, image5pos[1]:image5pos[1] + width_offset]


image_downsampled = cv2.resize(image, (image.shape[1]//2, image.shape[0]//2))

image1_downsampled = image_downsampled[image1pos_downsampled[0]:image1pos_downsampled[0] + height_offset//2, image1pos_downsampled[1]:image1pos_downsampled[1] + width_offset//2]
image2_downsampled = image_downsampled[image2pos_downsampled[0]:image2pos_downsampled[0] + height_offset//2, image2pos_downsampled[1]:image2pos_downsampled[1] + width_offset//2]
image3_downsampled = image_downsampled[image3pos_downsampled[0]:image3pos_downsampled[0] + height_offset//2, image3pos_downsampled[1]:image3pos_downsampled[1] + width_offset//2]
image4_downsampled = image_downsampled[image4pos_downsampled[0]:image4pos_downsampled[0] + height_offset//2, image4pos_downsampled[1]:image4pos_downsampled[1] + width_offset//2]
image5_downsampled = image_downsampled[image5pos_downsampled[0]:image5pos_downsampled[0] + height_offset//2, image5pos_downsampled[1]:image5pos_downsampled[1] + width_offset//2]
```
